Q.10. How can you tell the difference between break and continue?

Ans. Break statement moves the execution outside and just after a loop. 

Continue statement moves the execution to the start of the loop.

Example Code:

````python
print('Use of Break')
for i in range(10):
    if(i==7):
        break
    print(i)
    
print('Breaked')
print('\n')

print('Use of Continue')
for i in range(10):
    if(i==7):
        continue
    print(i)
````
