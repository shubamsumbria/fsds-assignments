```python
import pandas as pd
import matplotlib.pyplot as plt

# URL of the CSV file
csv_url = "https://data.wa.gov/api/views/f6w7-q2d2/rows.csv?accessType=DOWNLOAD"

# Read the CSV file into a DataFrame
data = pd.read_csv(csv_url)
data.head()
```

```python
# Question 1: Get all the cars and their types that do not qualify for clean alternative fuel vehicle
not_qualify_for_clean_alternative_fuel = data[data['Clean Alternative Fuel Vehicle (CAFV) Eligibility']!='Clean Alternative Fuel Vehicle Eligible']
not_qualify_for_clean_alternative_fuel.head()
```

```python
# Question 2: Get all TESLA cars with the model year, and model type made in Bothell City.
test_car_in_bothell_city = data[(data['City']=='Bothell') & (data['Make']=='TESLA')][['Model Year', 'Model']]
test_car_in_bothell_city.head()
```

```python
# Question 3: Get all the cars that have an electric range of more than 100, and were made after 2015
car_range_more_than_100_and_made_after_2015 = data[(data['Electric Range']>100) & (data['Model Year']>2015)]
car_range_more_than_100_and_made_after_2015.head()
```

```python
# Question 4: Draw plots to show the distribution between city and electric vehicle type
import seaborn as sns
import matplotlib.pyplot as plt
def plot_distribution_city_vehicle_type(df):
    fig, ax = plt.subplots(figsize=(12, 8))
    sns.countplot(x='City', hue='Electric Vehicle Type', data=df, ax=ax)
    plt.show()
    return fig, ax

plot_distribution_city_vehicle_type(data)
```
