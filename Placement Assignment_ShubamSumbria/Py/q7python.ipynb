```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the CSV file
data = pd.read_csv("earth_meteorites.csv")

# Convert the year column to datetime objects
data["year"] = pd.to_datetime(data["year"], format="%Y-%m-%dT%H:%M:%S.%f", errors='coerce')
```

```python
# Question 1: Get all the Earth meteorites that fell before the year 2000
earth_meteorites_before_2000 = data[data["year"].dt.year < 2000]
print("Earth meteorites fell before the year 2000:")
print(earth_meteorites_before_2000["name"].to_list())
```

```python
# Question 2: Get all the earth meteorites co-ordinates that fell before the year 1970
earth_meteorites_before_1970 = data[data["year"].dt.year < 1970]
print("Earth meteorites fell before the year 1970:")
print(earth_meteorites_before_1970["name"].to_list())
```

```python
# Question 3: Get all earth meteorites whose mass was more than 10000kg
earth_meteorites_mass_greater_than_10000 = data[data["mass"] > 10000]
print("Earth meteorites with mass greater than 10000kg:")
print(earth_meteorites_mass_greater_than_10000["name"].to_list())
```
