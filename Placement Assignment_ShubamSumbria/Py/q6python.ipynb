```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime, timedelta

# Load the data from the CSV file
data = pd.read_excel("pokemon_output.xlsx")

# Convert spawn_chance column to numeric
data["spawn_chance"] = pd.to_numeric(data["spawn_chance"], errors="coerce")
```

```python
# Question 1: Get all Pokemons whose spawn rate is less than 5%
spawn_rate_threshold = 5
low_spawn_rate_pokemons = data[data["spawn_chance"] < spawn_rate_threshold]
print("Pokemons with spawn rate less than 5%:")
print(low_spawn_rate_pokemons["name"].to_list())

# Plot spawn rate distribution
plt.hist(data["spawn_chance"], bins=20, edgecolor="black")
plt.xlabel("Spawn Rate")
plt.ylabel("Frequency")
plt.title("Distribution of Spawn Rate")
plt.show()
```

```python
# Question 2: Get all Pokemons that have less than 4 weaknesses
max_weaknesses = 4
less_weakness_pokemons = data[data["weakness"].apply(lambda x: len(x.split(","))) < max_weaknesses]
print("Pokemons with less than 4 weaknesses: ", less_weakness_pokemons["name"].to_list())

# Plot weaknesses distribution
data["num_weaknesses"] = data["weakness"].apply(lambda x: len(x.split(",")))
plt.hist(data["num_weaknesses"], bins=range(max_weaknesses+1), edgecolor="black")
plt.xlabel("Number of Weaknesses")
plt.ylabel("Frequency")
plt.title("Distribution of Number of Weaknesses")
plt.xticks(range(max_weaknesses+1))
plt.show()
```

```python
# Question 3: Get all Pokemons that have no multipliers at all
no_multipliers_pokemons = data[data["multipliers"].isnull()]
print("Pokemons with no multipliers:")
print(no_multipliers_pokemons["name"].tolist())
```

```python
# Question 4: Get all Pokemons that do not have more than 2 evolutions
max_evolutions = 2
less_evolutions_pokemons = data[data["next_evolution"].apply(lambda x: len(x.split(",")) if pd.notnull(x) else 0) <= max_evolutions]
print("Pokemons with no more than 2 evolutions:")
print(less_evolutions_pokemons["name"].tolist())

# Plot evolution count distribution
data["num_evolutions"] = data["next_evolution"].apply(lambda x: len(x.split(",")) if pd.notnull(x) else 0)
plt.hist(data["num_evolutions"], bins=range(max_evolutions+2), edgecolor="black")
plt.xlabel("Number of Evolutions")
plt.ylabel("Frequency")
plt.title("Distribution of Number of Evolutions")
plt.xticks(range(max_evolutions+2))
plt.show()
```

```python
# Convert spawn_time column to datetime objects
data["spawn_time"] = pd.to_datetime(data["spawn_time"], format="%M:%S")

# Question 5: Get all Pokemons whose spawn time is less than 300 seconds
max_spawn_time = timedelta(seconds=300)
reference_datetime = datetime(1970, 1, 1)  # Use a fixed reference datetime
max_spawn_datetime = reference_datetime + max_spawn_time
less_spawn_time_pokemons = data[data["spawn_time"] < max_spawn_datetime]
print("Pokemons with spawn time less than 300 seconds:")
print(less_spawn_time_pokemons["name"].tolist())

# Calculate spawn time in seconds
data["spawn_seconds"] = data["spawn_time"].dt.minute * 60 + data["spawn_time"].dt.second

# Plot spawn time distribution
plt.hist(data["spawn_seconds"], bins=20, edgecolor="black")
plt.xlabel("Spawn Time (Seconds)")
plt.ylabel("Frequency")
plt.title("Distribution of Spawn Time")
plt.show()
```
